```python
terr = pd.read_csv('/home/pratyush/Downloads/globalterrorismdb_0919dist.csv')
terr.head()
```

```python
terr = terr[['iyear' ,'country' ,'country_txt' , 'nkill']].fillna(value = 0)
terr.info() #selecting relevant columns
```

```python
terr_1 = terr.groupby(['country' , 'country_txt','iyear' ] , as_index = False).sum().sort_values(by = ['country_txt','iyear'])
terr_1.reset_index(drop= True , inplace = True)
terr_1.head() 
#groupby to find country vise yearly deaths and then reset index
```

```python
terr_1.rename(columns = {'iyear' : 'Year' , 'nkill' : 'deaths'},inplace = True)
terr_1.info() 
#change column names
```
